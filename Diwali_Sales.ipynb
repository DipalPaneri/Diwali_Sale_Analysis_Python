# Diwali Sales Analysis - Explaratory Data Analysis Project in Python 

```python
#Import all the libraries will be using for this Project
#Matplotlib inline makes the visualization more effective directly in cell. 

import numpy as np
import pandas as pd #Data Cleaning
import matplotlib.pyplot as plt #Visualization Library 
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\paner\Documents\Pandas\Diwali Sales Data.csv", encoding= 'unicode_escape') 
#Added r to mark it as a raw string helps with special characters, to avoid encoding error used 'unicode_escape'
```

```python
#shape method gives us the total no of rows and columns in the data table.
df.shape
```

```python
#head() method will show first 5 rows of data with all 15 columns. 
df.head(10)
```

#### Python Data Cleaning

```python
#Usually Data Cleaning can be done with other tools i.e SQL Or Power BI, However,These are the following menthods to do the same with python.
df.info()
```

```python
#to delete the records or empty columns drop() function is used
#axis parameter is used to delete the entire column from the database table. 
#inplace True is para. used to commit the changes or save the changes we made with drop function.

df.drop(['Status', 'unnamed1'], axis=1, inplace=True )
```

```python
#check if the Columns are deleted.

df.info()
```

```python
#Now We will check the null values
#pd.isnull(df), will give the result in False, so the entire databse will look messy with 'False'.
#adding sum() will get us the exact no. of null or empty records in the databse and in which column it is. 

pd.isnull(df).sum()
```

```python
#dropna() is used to delete those nul value records. 

df.dropna(inplace=True)
```

```python
df.shape

#result shows 11251 rows reduced to 11239
```

```python
#Changing the Data Type of certain Columns
#astype(int/float/complex/string) used to change the datatype of the column.


df['Amount'] = df['Amount'].astype(int)
```

```python
df['Amount'].dtype
```

```python
#columns to fetch the names of database table columns
df.columns
```

```python
#rename the column name, It will not save the changes as inplace para. was not passed.
df.rename(columns = {'Marital_Status' : 'Status'})
```

```python
#describe() method is used to describe the numerical details of DataFrame exp. count, mean, min, max. 
df.describe()
```

```python
#some of the details given in describe() method as 50% 25% in some columns are useless and extra memory which we dont need.
df[['Age', 'Orders', 'Amount']].describe()
```

## Exploratory Data Analysis

GENDER

```python
#Plotting a bar chart for Gender and its Count using Seaborn
#for the value of exact count of F and M, we add for loop condition

ax = sns.countplot(x='Gender', data = df)
sns.set(rc={'figure.figsize': (9, 6)})  

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#plotting a bar chart of Gender vs total Amount

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)

sns.barplot (x = 'Gender',y = 'Amount' ,data = sales_gen)
```

From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

## Age 

```python
#Plot a bar chart using Age Group vs Amount with respect to Gender 
#hue para. works similar to legends in Pivottables in Excel. As a detailed categorized filter

ax = sns.countplot(x='Age Group', hue ='Gender', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Ploting a Bar Chart of Age Group vs Amount

sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group', y='Amount', data=sales_age)
```

From Above Graphs, we can see most buyers are from age group between 26-35 females.

### State 

```python
df.columns
```

```python
#Now We will plot a bar chart for State vs Orders 

sales_state=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(13)
sns.set(rc={'figure.figsize': (18,7)})
sns.barplot(x='State', y='Orders', data=sales_state)
```

```python
#Total Amount of Sales per state 

sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(14)
sns.set(rc={'figure.figsize': (19,6)})
sns.barplot(x='State', y='Amount', data=sales_state)
```

From the Above Graphs, we can see that most of the orders were received from Utter Pradesh, Maharashta,Karnataka and Delhi respectively. 

## Sales Analysis based on Marital Status

```python
ax = sns.countplot(x='Marital_Status', data=df)
sns.set(rc={'figure.figsize': (9,7)})

for bar in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x='Marital_Status', y='Amount', hue = 'Gender', data=sales_state)
```

From the above Graph, we can see most of the customers are Married Women who has high purchasing capabilities. 

## Sales Analysis based on Occupation

```python
ax = sns.countplot(x='Occupation', data=df)
sns.set(rc={'figure.figsize': (18,5)})

for bar in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Occupation', y='Amount', data=sales_state)
```

## Sales Analysis based on Product Category

```python
ax = sns.countplot(x ='Product_Category', data=df)
sns.set(rc={'figure.figsize':(16, 6)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(15)
sns.set(rc={'figure.figsize': (25,7)})
sns.barplot(x='Product_Category', y='Amount', data=sales_state)

for bars in ax.containers:
    ax.bar_label(bars)
```

From the above graph, we can see most of the spending happened on food, Clothing, Electronics and Footwear.
